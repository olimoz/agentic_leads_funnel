```python
def get_price(model, date, price_type):
    """
    Get the applicable price for a given model, date, and price type.
    
    Args:
    model (str): The model name.
    date (datetime): The date for which to get the price.
    price_type (str): The type of price to retrieve (e.g., 'input_price').
    
    Returns:
    float: The applicable price, or None if no price is found.
    """

    model_prices = df_prices[df_prices['model'] == model]
    valid_prices = model_prices[
        (model_prices['start_date'] <= date) & 
        ((model_prices['end_date'] >= date) | (model_prices['end_date'].isna()))
    ]
    
    if len(valid_prices) == 0:
        previous_prices = model_prices[model_prices['start_date'] <= date]
        if len(previous_prices) > 0:
            return previous_prices.iloc[-1][price_type]
        return None
    
    return valid_prices.iloc[-1][price_type]
```

```python
def calculate_price_columns(df):
    """
    Calculate total search price and token price columns for a DataFrame.
    
    Parameters:
    df (pandas.DataFrame): DataFrame containing columns:
        - search_calls
        - search_price
        - input_tokens
        - input_price
        - output_tokens
        - output_price
    
    Returns:
    pandas.DataFrame: Original DataFrame with three new columns:
        - search_price_tot: (search_calls * search_price) / 1000000
        - token_price_tot: ((input_tokens * input_price) + (output_tokens * output_price)) / 1000000
        - total_price: search_price_tot + token_price_tot
    """
    # Create a copy to avoid modifying the original DataFrame
    result_df = df.copy()
    
    # Calculate total search price
    result_df['search_price_tot'] = result_df.apply(
        lambda row: (row['search_calls'] * row['search_price'])
        if pd.notna(row['search_calls']) and pd.notna(row['search_price']) 
        else np.nan, 
        axis=1
    )
    
    # Calculate total token price
    result_df['token_price_tot'] = result_df.apply(
        lambda row: (
            ((row['input_tokens'] * row['input_price']) / 1000000 if pd.notna(row['input_tokens']) and pd.notna(row['input_price']) else 0) +
            ((row['output_tokens'] * row['output_price']) / 1000000 if pd.notna(row['output_tokens']) and pd.notna(row['output_price']) else 0)
        ),
        axis=1
    )
    
    # If both components were NaN, set the total to NaN instead of 0
    result_df.loc[result_df.apply(lambda row: 
        (pd.isna(row['input_tokens']) or pd.isna(row['input_price'])) and 
        (pd.isna(row['output_tokens']) or pd.isna(row['output_price'])), 
        axis=1), 'token_price_tot'] = np.nan
    
    # Calculate total price (sum of search and token prices)
    result_df['total_price'] = result_df.apply(
        lambda row: (
            (row['search_price_tot'] if pd.notna(row['search_price_tot']) else 0) +
            (row['token_price_tot'] if pd.notna(row['token_price_tot']) else 0)
        ),
        axis=1
    )
    
    # If both components were NaN, set the total price to NaN
    result_df.loc[
        result_df['search_price_tot'].isna() & result_df['token_price_tot'].isna(), 
        'total_price'
    ] = np.nan

    return result_df

# Usage example:
# df_with_totals = calculate_price_columns(df)
```

```python
import pandas as pd
import re

def extract_blog_urls(text, exclusion_list=None):
    if exclusion_list is None:
        exclusion_list = ['facebook.com', 'linkedin.com']
    
    # Modified pattern to look for 'blog' in the URL
    url_pattern = r'\b(?:https?:\/\/)?(?:www\.)?([a-zA-Z0-9-]+(?:\.[a-zA-Z0-9-]+)+(?:\/\S*blog\S*)?)'
    
    urls = re.findall(url_pattern, text, re.IGNORECASE)
    
    clean_urls = []
    for domain in urls:
        if domain.startswith('www.'):
            domain = domain[4:]
        
        if 'blog' in domain.lower() and not any(excluded_domain in domain for excluded_domain in exclusion_list):
            clean_urls.append(domain)
    
    return list(dict.fromkeys(clean_urls))
```

```python

class TargetScoreAgent():

    def run(self, llm, prompt_text, params):

        # Define tools for scoring
        tools = [{
            "name": "score_business",
            "description": "Score of business text for video production need, as per metrics",
            "parameters": {
                "type": "object",
                "properties": {
                    "reasoning": {
                        "type": "string",
                        "description": "Explanation for the score given the metrics"
                    },                        
                "video_need_score": {
                        "type": "integer",
                        "description": "Score of the text to inform video production needs, as per metrics (0-10)"
                    },
                "summary_of_facts": {
                        "type": "string",
                        "description": "List of facts in the text which led to the score, mentioning sources wherever possible but not metrics. List source URLs at end."
                    }
                },
                "required": ["reasoning", "video_need_score", "summary_of_facts_and_urls", "list_of_URLs"]
            }
        }]

        # Create the prompt using ChatPromptTemplate
        prompt = ChatPromptTemplate.from_messages([
            ("system", "You score results of web searches according to a strict set of metrics"),
            ("human", prompt_text),
        ]).partial()

        # Create the chain to invoke the language model
        prompt_input = {
            'first_name': lambda x: x['first_name'],
            'last_name': lambda x: x['last_name'],
            'company': lambda x: x['company'],
            'search_period': lambda x: x['search_period'],
            'details': lambda x: x['details']
        }
        chain = (prompt_input | prompt | llm.bind_tools(tools, tool_choice="score_business"))

        # Invoke the chain asynchronously
        response = chain.invoke(params)

        # print(response)

        # Extract function arguments from the response
        func_args = response.additional_kwargs['tool_calls'][0]['function']['arguments']

        # Extract the scores from the JSON response
        try:
            func_args_data = json.loads(func_args)
            video_need_score = int(func_args_data['video_need_score'])
            reasoning = func_args_data['reasoning']
            summary = func_args_data['summary_of_facts']

            input_tokens=response.response_metadata['token_usage']['prompt_tokens']
            output_tokens=response.response_metadata['token_usage']['completion_tokens']

        except Exception as e:
            msg = f"TargetScoreAgent: {str(e)}"
            print(msg)
            video_need_score = 0
            reasoning = ""
            summary = ""
            input_tokens= 0
            output_tokens=0

        # Return the extracted scores and reasoning
        return video_need_score, reasoning, summary, input_tokens, output_tokens
```
